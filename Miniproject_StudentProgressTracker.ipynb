**STUDENT PROGRESS TRACKER SYSTEM**

```python
program = True                                         #Control main loop
while program:                                              #while true is an infinite loop that repeats until the user exits.

  print("STUDENT PROGRESS TRACKER SYSTEM")

  #Get student details from the user.

  STUDENT_NAME=input("Enter Student Name:")
  CLASS=input("Enter Class:")
  ROLL_NUMBER=input("Enter Roll Number:")
  SUBJECTS = int(input("Enter no.of subjects:"))

  #Set fail as False and total marks as 0.

  fail=False
  total = 0

  #Read marks for each subject.

  for i in range(SUBJECTS):
    marks = int(input(f"Enter subject {i+1} marks: "))
    total += marks                                       #add total=total+marks
    print("SUBJECT", i+1, "=", marks)

    #If marks are below 35, student fails in that subject.

    if marks < 35:
        fail = True
  average = total / SUBJECTS                            #Find the average marks.
  print("\nTotal Marks =", total)
  print("Average Marks =", average)

  #Check average and assign grade.

  if(average>=90):
    grade = "A grade"
  elif(average>75)&(average<89):
    grade = "B grade"
  elif(average>50)&(average<74):
    grade = "C grade"
  elif(average>35)&(average<49):
    grade = "D grade"
  else:
    grade = "fail"

    #Decide pass or fail.

  if fail:
    result = "Fail"
  else:
    result = "Pass"

  while True:                     #report main loop

    #Show the full student report

    print("\n=========== STUDENT PROGRESS REPORT ===========")
    print(f"{'Name:':15} {STUDENT_NAME}")
    print(f"{'Class:':15} {CLASS}")
    print(f"{'Roll Number:':15} {ROLL_NUMBER}")
    print(f"{'Total Marks:':15} {total}")
    print(f"{'Average Marks:':15} {average:.2f}")
    print(f"{'Grade:':15} {grade}")
    print(f"{'Result:':15} {result}")
    print("================================================")

  #Ask the user what to do next.

    print("\nWhat do you want to do next?")
    print("1. Enter marks again")
    print("2. View report again")
    print("3. Exit")
    choice = int(input("Enter your choice: "))
    if choice == 1:
      break                               #Exit report loop â†’ go back to main loop

    elif choice == 2:
      continue                            #Show report again if user selects 2.
    elif choice == 3:
      print("Exiting... Thank you!")
      program = False
      break                               #Exit the program if user selects 3.

    #If the user enters an invalid option, show error message.
    else:
      print("Invalid choice! Try again.")
```
